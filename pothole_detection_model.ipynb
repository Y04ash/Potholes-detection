```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("rajdalsaniya/pothole-detection-dataset")

print("Path to dataset files:", path)
```

```python
from ultralytics import YOLO

# Load a pre-trained YOLOv8 model
model = YOLO("yolov8n.pt")  # 'n' stands for nano version (fast and small)

# Run inference on an image
# results = model("road2.jpg", save=True)  # Detect objects in the image and save results
```

```python
# Train YOLOv8 on your dataset
data_path = path + "/data.yaml"
model.train(data=data_path, epochs=10, imgsz=640)
```

```python
metrics = model.val()
```

```python
import matplotlib.pyplot as plt
from PIL import Image
results = model("pothole3.webp")
# Assuming 'results' is the output from your model prediction
# and it contains the image data you want to display.
results[0].plot()
# Display the image using matplotlib.
# plt.imshow(image)
# plt.axis('off')  # Hide the axis ticks and labels.
# plt.show()
```
